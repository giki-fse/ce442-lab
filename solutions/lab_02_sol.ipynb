$$f(x) = \cos(x)$$

$$cos(x) = \sum_{k=0}^{\infty} \frac{(-1)^k x^{2k}}{(2k)!}$$

```python
import numpy as np;
import matplotlib.pyplot as plt;
import math;

def compute_taylor_series(x,order):

    approx = np.zeros_like(x);
    for k in range((order//2)+1):
        approx += ((-1)**k * x **(2*k))/ (math.factorial(2*k))
    
    return approx

x = np.linspace(-np.pi, np.pi, 200)

test_val = 1.5
true_val = np.cos(test_val)

orders = [0, 2, 4, 6]

test_value_approximations = {}
test_value_errors = {}

plt.figure(figsize=(10,8))

for order in orders:
    y = compute_taylor_series(x,order)
    
    # plot
    plt.plot(x,y,label= f'{order} Order')

    val = compute_taylor_series(test_val, order)
    test_value_approximations[order] = val
    test_value_errors[order] = abs(true_val - val)

# Plot the actual cos function
plt.plot(x, np.cos(x), 'k', label='Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

plt.show()

print(f'True value of sin(1.5): {true_val:.6f}')
for order in orders:
    print(f'Value for {order} Order Taylor series approximation at x = {test_val}: {test_value_approximations[order]:.6f}')
    print(f'Error for {order} Order Taylor series approximation at x = {test_val}: {test_value_errors[order]:.6f}')

```
